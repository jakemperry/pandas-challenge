```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Count or get average values from data frames that do not need to account for specific indexes
distSchools = school_data['School ID'].count()
distStudents = student_data['Student ID'].count()
distBudget = school_data['budget'].sum()
distAvgMathScore = school_data_complete['math_score'].mean()
distAvgReadingScore = school_data_complete['reading_score'].mean()

distPassMathStudents= school_data_complete.loc[school_data_complete['math_score']>=70]
distPassMathPct = (distPassMathStudents['Student ID'].count()/distStudents)*100

distPassReadingStudents = school_data_complete.loc[school_data_complete['reading_score']>=70]
distPassReadingPct = (distPassReadingStudents['Student ID'].count()/distStudents)*100

distPassBothStudents = pd.merge(distPassMathStudents,distPassReadingStudents, on=['Student ID','student_name','gender','grade','school_name','School ID','type','size','budget'])
distPassBothPct = (distPassBothStudents['Student ID'].count()/distStudents)*100

#print(f'{distSchools}\n{distStudents}\n{distBudget}\n{distAvgMathScore}\n{distAvgReadingScore}\n{distPassMathPct}\n{distPassReadingPct}\n{distPassBothPct}')

distSummary = pd.DataFrame({
    'TotalSchools':distSchools,
    'TotalStudents':distStudents,
    'TotalBudget':distBudget,
    'AvgMathScore':distAvgMathScore,
    'AvgReadingScore':distAvgReadingScore,
    '%PassingMath':distPassMathPct,
    '%PassingReading':distPassReadingPct,
    '%PassingOverall':distPassBothPct,},
    index=[0]
)
distSummary.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name

  * Total School Budget

  * Average Math Score

```python
schoolGroup = school_data_complete.groupby(['school_name'])

schoolGroupType = schoolGroup['type'].unique()
schoolGroupStudents = schoolGroup['Student ID'].count()
schoolGroupBudget = schoolGroup['budget'].unique().astype(float)
schoolGroupPerStudentBudget = schoolGroupBudget/schoolGroupStudents
schoolGroupAvgMath = schoolGroup['reading_score'].mean()
schoolGroupAvgReading = schoolGroup['math_score'].mean()

mathPassGroup = distPassMathStudents.groupby(['school_name'])
schoolGroupMathPass = mathPassGroup['Student ID'].count()/schoolGroupStudents*100

readingPassGroup = distPassReadingStudents.groupby(['school_name'])
schoolGroupReadingPass = readingPassGroup['Student ID'].count()/schoolGroupStudents*100

passBothGroup = distPassBothStudents.groupby(['school_name'])
schoolGroupPassBoth = passBothGroup['Student ID'].count()/schoolGroupStudents*100

schoolSummary = pd.DataFrame({
    'SchoolType':schoolGroupType,
    'TotalStudents':schoolGroupStudents,
    'TotalSchoolBudget':schoolGroupBudget,
    'PerStudentBudget':schoolGroupPerStudentBudget,
    'AvgMathScore':schoolGroupAvgMath,
    'AvgReadingScore':schoolGroupAvgReading,
    '%PassingMath':schoolGroupMathPass,
    '%PassingReading':schoolGroupReadingPass,
    '%OverallPassing':schoolGroupPassBoth
})
schoolSummary
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Make a dataframe that's a subset of the complete school data
mathGrades = school_data_complete[['school_name','grade','math_score']]
#Make individual dataframes for each grade 
mathNine = mathGrades.loc[mathGrades['grade']=='9th']
mathTen = mathGrades.loc[mathGrades['grade']=='10th']
mathEleven = mathGrades.loc[mathGrades['grade']=='11th']
mathTwelve = mathGrades.loc[mathGrades['grade']=='12th']
#Group by school name
mathNineGroup = mathNine.groupby(['school_name'])
mathTenGroup = mathTen.groupby(['school_name'])
mathElevenGroup = mathEleven.groupby(['school_name'])
mathTwelveGroup = mathTwelve.groupby(['school_name'])
#Calculate average grades for each school
mathNineAvg = mathNineGroup['math_score'].mean()
mathTenAvg = mathTenGroup['math_score'].mean()
mathElevenAvg = mathElevenGroup['math_score'].mean()
mathTwelveAvg = mathTwelveGroup['math_score'].mean()
#Put the dataframes together into a single dataframe
mathGradeAvgs1 = pd.merge(mathNineAvg, mathTenAvg, on='school_name', suffixes=('_9','_10'))
mathGradeAvgs2 = pd.merge(mathElevenAvg,mathTwelveAvg, on='school_name', suffixes=('_11','_12'))
mathGradeAvgs = pd.merge(mathGradeAvgs1,mathGradeAvgs2, on='school_name')
#Rename the columns
mathGradeAvgs = mathGradeAvgs.rename(columns={'math_score_9':'9th Grade',
                                              'math_score_10':'10th Grade',
                                              'math_score_11':'11th Grade',
                                              'math_score_12':'12th Grade'
                                             })
mathGradeAvgs
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Set up a new dataframe copy and binning
scoreSpending=school_data_complete
scoreSpending['PerStudentBudget']=scoreSpending['budget']/scoreSpending['size']
bins = [575.0,584.9,629.9,644.9,680.0]
groupNames = [f"<$585","${}".format(585)+"-630","${}".format(630)+"-645","${}".format(645)+"-680"]
scoreSpending["PerStudentBudgetRange"]= pd.cut(scoreSpending["PerStudentBudget"], bins, labels=groupNames, include_lowest=True)

#Create groupby for the new bins
scoreGroups = scoreSpending.groupby(['PerStudentBudgetRange'])
scoreMathAvg = scoreGroups['math_score'].mean()
scoreReadAvg = scoreGroups['reading_score'].mean()
scoreMathPass = scoreGroups['math_score'].apply(lambda x: x[x>=70].count())
scoreMathCount = scoreGroups['math_score'].count()
scoreMathPass = (scoreMathPass/scoreMathCount)*100
scoreReadPass = scoreGroups['reading_score'].apply(lambda x: x[x>=70].count())
scoreReadCount = scoreGroups['reading_score'].count()
scoreReadPass = (scoreReadPass/scoreReadCount)*100

#Overall passing needs to use math passing AND reading passing students
#Find all math passing students
scoreMathPass1 =scoreSpending.loc[scoreSpending['math_score']>=70]
scoreMathPass1 = scoreMathPass1.drop(columns=['gender','grade','school_name','reading_score','School ID','type','size','budget','PerStudentBudget'])
scoreMathPass1 = scoreMathPass1.set_index('Student ID')
#Find all reading passing students
scoreReadPass1 =scoreSpending.loc[scoreSpending['reading_score']>=70]
scoreReadPass1 = scoreReadPass1.drop(columns=['gender','grade','school_name','math_score','School ID','type','size','budget','PerStudentBudget'])
scoreReadPass1 = scoreReadPass1.set_index('Student ID')
#Merge on inner join to find unique students who pass both reading and math
scoreBothPass = scoreMathPass1.join(scoreReadPass1,on='Student ID',how='inner', lsuffix='_math',rsuffix='_read')
```

```python
#Create a groupby to apply the bins to the merged passing student data
scoreBothGroups = scoreBothPass.groupby(['PerStudentBudgetRange_math'])
scoreBothPass1 = scoreBothGroups['math_score'].count()
scoreBothPass1 = (scoreBothPass1/scoreReadCount)*100

#Put it all together into a summary dataframe
scoreSpendingSummary = pd.DataFrame({
    'Average Math Score':scoreMathAvg.map("{:.2f}".format),
    'Average Reading Score':scoreReadAvg.map("{:.2f}".format),
    '%PassingMath':scoreMathPass.map("{:.2f}".format),
    '%PassingReading':scoreReadPass.map("{:.2f}".format),
    '%PassingOverall':scoreBothPass1.map("{:.2f}".format)
})
scoreSpendingSummary.index.rename('SpendRangePerStudent')
scoreSpendingSummary
```
